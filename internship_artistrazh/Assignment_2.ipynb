````python
class ListNode:
        def __init__(self, value, next_):
            self.value = value
            self.next = next_
        def print_all(self):  # –º–µ—Ç–æ–¥ –¥–ª—è —É–¥–æ–±—Å—Ç–≤–∞
            while self != None:
                print(self.value)
                self = self.next

# –∞–ª–≥–æ—Ä–∏—Ç–º —Ä–∞–±–æ—Ç–∞–µ—Ç –∑–∞ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –≤—Ä–µ–º—è
# –±–µ–∑ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –≤—Å—Ç—Ä–æ–µ–Ω–Ω—ã—Ö —Å—Ç—Ä—É–∫—Ç—É—Ä –¥–∞–Ω–Ω—ã—Ö
    
def my_function(head):
    node = head
    reverse_list = None
    list_len = 0
    while node != None:
        list_len += 1
        node = node.next
    
    import math 
    for i in range(math.floor(list_len/2.)):
        node = head
        for j in range(i):
            node = node.next
        
        curr_val_start = node.value

        node = head
        for k in range(list_len-i-1):
            node = node.next
            
        curr_val_end = node.value
        
        if curr_val_start != curr_val_end:
            return False
        
    return True
            
````

````python
list_head = ListNode(1, ListNode(2, ListNode(2, ListNode(3, None))))
list_head.print_all()
my_function(list_head)
````
